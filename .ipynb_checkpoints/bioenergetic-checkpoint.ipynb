```julia
using BioEnergeticFoodWebs
```

```julia
using Plots
```

```julia
using DataFrames, CSV
```

```julia
s_vec = [10, 20, 25, 50, 100]
c_vec = [0.1, 0.15, 0.2, 0.3]
df = DataFrame(S=[], C=[], survS=[])
n_reps = 10
```

```julia
for S in s_vec
    for C in c_vec
        for R in n_reps
            A = nichemodel(S, C)
            parameters = model_parameters(A)
            bm = rand(Float64, S).*(1/S)
            @time sim = simulate(parameters, bm; stop=500, use=:stiff)
            t_max = maximum(sim[:t])
            surv_species= sum(sim[:B][Int64(t_max),:].>0.0)
            
            push!(df.S, S)
            push!(df.C, C)
            push!(df.survS, surv_species)
        end
    end
end
CSV.write("testout.csv", df)
```

```julia
@time sim = simulate(parameters, bm; stop=500, use=:stiff)
minimum(sim[:B])

pl2 = vec(sum(sim[:B].>0.0, dims = 2)) |> x -> plot(sim[:t], x, leg=false)
pl1 = plot(sim[:t], sim[:B], leg=false, c=:teal)
plot(pl1, pl2)
```
