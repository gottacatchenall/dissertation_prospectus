```julia
using DifferentialEquations
using Plots
```

```julia
function macarthur_consumer_resource(dB, B, p, t)
    R, C = B
    attack_rate_RC, weight_C, metabolic_demand_C, B_i, G_R = p
    dB[1] = dR = R* (G_R*(1.0-R) - attack_rate_RC * C)
    dB[2] = dC = C * B_i * (attack_rate_RC * weight_C * R - metabolic_demand_C)
end

B0 = [1.0, 0.2]
params = (1.0, 1.0, 2.0, 1.0, 1.0)
tspan = (0.0, 10.0)
prob = ODEProblem(macarthur_consumer_resource, B0, tspan, params)
```

```julia
sol = solve(prob)
```

```julia
plot(sol)
```
